```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.linspace(0, 9, 10)
```

```python
x = np.linspace(1, 10, 1000)
y1 = np.power(x, 0.5)
y2 = 2 * np.sin(x)
y3 = 3 * np.cos(x)
```

```python
plt.plot(x, y1, label='$\sqrt{x}$')
plt.plot(x, y2, label='$2 \dot \sin{x}$')
plt.plot(x, y3, 'b--', label='$3 \dot \cos{x}$')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("comparison of functions")
# plt.xlim(0, 5)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 2), dpi=300)
ax1, ax2 = axes
ax1.plot(x, y1)
ax2.plot(x, y2)
ax1.set_xlabel('x')
ax2.set_ylabel('f(x)')
fig.suptitle('graphics')
```

```python
names = ['a', 'b', 'c']
values = [100, 150, 180]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10), dpi=300)
(ax1, ax2), (ax3, ax4) = axes

ax1.bar(names, values, color=['r', 'g', 'y'])
ax1.set_title('bar')
ax2.scatter(names, values)
ax2.set_title('scatter')
ax3.plot(names, values)
ax3.set_title('plot')
ax4.pie(values)
ax4.set_title('pie')
```
