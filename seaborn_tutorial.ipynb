```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head(100)
```

```python
ax = sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm')
```

```python
fg = sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='scatter')
```

```python
sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='scatter', col='sex', hue='species')
```

```python
sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', kind='scatter', col='sex', hue='island')
```

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
sns.relplot(data=flights, y='passengers', x='year', kind='line')
```

```python
sns.displot(data=penguins, x='body_mass_g', kind='hist', bins=20)
```

```python
sns.displot(data=penguins, x='body_mass_g', kind='hist', bins=20, hue='island')
```

```python
sns.displot(data=penguins, x='body_mass_g', kind='hist', bins=20, hue='species', col='sex', row='island')
```

```python
sns.displot(data=penguins, x='body_mass_g', kind='kde')
```

```python
sns.displot(data=penguins, x='body_mass_g', y='bill_length_mm', cbar=True)
```
